## 데이터 구조 확인

```python
import pandas as pd
import numpy as np
from pathlib import Path
from IPython.display import display

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 100)
```

```python
## 경로 설정 셀
NOTEBOOK_DIR = Path.cwd()
PROJECT_ROOT = NOTEBOOK_DIR.parent
DATA_DIR = PROJECT_ROOT / "data"

print("NOTEBOOK_DIR:", NOTEBOOK_DIR)
print("PROJECT_ROOT:", PROJECT_ROOT)
print("DATA_DIR:", DATA_DIR)
```

```python
## 데이터 로드 셀
train = pd.read_csv(DATA_DIR / "train.csv")
client = pd.read_csv(DATA_DIR / "client.csv")
electricity_prices = pd.read_csv(DATA_DIR / "electricity_prices.csv")
gas_prices = pd.read_csv(DATA_DIR / "gas_prices.csv")
forecast_weather = pd.read_csv(DATA_DIR / "forecast_weather.csv")
historical_weather = pd.read_csv(DATA_DIR / "historical_weather.csv")

print("train:", train.shape)
print("client:", client.shape)
print("electricity_prices:", electricity_prices.shape)
print("gas_prices:", gas_prices.shape)
print("forecast_weather:", forecast_weather.shape)
print("historical_weather:", historical_weather.shape)
```

```python
## 컬럼 구조 확인 
display(train.head())
display(client.head())
display(electricity_prices.head())
display(gas_prices.head())
display(forecast_weather.head())
display(historical_weather.head())
```

```python
## 컬럼명 보기 
print("train columns:", train.columns.tolist())
print("client columns:", client.columns.tolist())
print("electricity_prices columns:", electricity_prices.columns.tolist())
print("gas_prices columns:", gas_prices.columns.tolist())
print("forecast_weather columns:", forecast_weather.columns.tolist())
print("historical_weather columns:", historical_weather.columns.tolist())
```

```python
## 타입 보기
train.info()
client.info()
electricity_prices.info()
gas_prices.info()
forecast_weather.info()
historical_weather.info()
```

```python
## 결측치 보기
def missing_ratio(df):
    return (df.isnull().mean() * 100).sort_values(ascending=False)

display(missing_ratio(train).head(20))
display(missing_ratio(client).head(20))
display(missing_ratio(forecast_weather).head(20))
display(missing_ratio(historical_weather).head(20))
```

```python
## datetime 변환
if "datetime" in train.columns:
    train["datetime"] = pd.to_datetime(train["datetime"])

if "date" in client.columns:
    client["date"] = pd.to_datetime(client["date"])

if "forecast_date" in electricity_prices.columns:
    electricity_prices["forecast_date"] = pd.to_datetime(electricity_prices["forecast_date"])

if "forecast_date" in gas_prices.columns:
    gas_prices["forecast_date"] = pd.to_datetime(gas_prices["forecast_date"])

if "forecast_datetime" in forecast_weather.columns:
    forecast_weather["forecast_datetime"] = pd.to_datetime(forecast_weather["forecast_datetime"])

if "datetime" in historical_weather.columns:
    historical_weather["datetime"] = pd.to_datetime(historical_weather["datetime"])

print(train.dtypes)
```

```python
## 기본 정렬
train = train.sort_values(["datetime"]).reset_index(drop=True)
historical_weather = historical_weather.sort_values(["datetime"]).reset_index(drop=True)
forecast_weather = forecast_weather.sort_values(["forecast_datetime"]).reset_index(drop=True)
```

```python
## 중복 확인
print("train duplicates:", train.duplicated().sum())
print("client duplicates:", client.duplicated().sum())
print("electricity_prices duplicates:", electricity_prices.duplicated().sum())
print("gas_prices duplicates:", gas_prices.duplicated().sum())
print("forecast_weather duplicates:", forecast_weather.duplicated().sum())
print("historical_weather duplicates:", historical_weather.duplicated().sum())
```

```python
## 생산 / 소비 분리 확인 1
train["is_consumption"].value_counts(dropna=False)
```

```python
## 생산 / 소비 분리 확인 2
train_cons = train[train["is_consumption"] == 1].copy()
train_prod = train[train["is_consumption"] == 0].copy()

print("consumption shape:", train_cons.shape)
print("production shape:", train_prod.shape)

display(train_cons["target"].describe())
display(train_prod["target"].describe())
```

```python
## train key 구조 확인

print("train columns:")
print(train.columns.tolist())

key_cols = ["county", "product_type", "is_business", "is_consumption", "datetime"]
existing_key_cols = [col for col in key_cols if col in train.columns]

print("\n사용 가능한 key 후보:")
print(existing_key_cols)

print("\n각 key 컬럼의 고유값 개수:")
for col in existing_key_cols:
    print(f"{col}: {train[col].nunique()}")

print("\nkey 후보 샘플:")
display(train[existing_key_cols].head())
```

```python
## client merge 방식 확인

print("client columns:")
print(client.columns.tolist())

client_key_candidates = ["county", "product_type", "is_business", "date", "installed_capacity", "eic_count"]
existing_client_cols = [col for col in client_key_candidates if col in client.columns]

print("\nclient에서 확인할 주요 컬럼:")
print(existing_client_cols)

display(client[existing_client_cols].head())

print("\nclient 고유 조합 개수:")
merge_check_cols = [col for col in ["county", "product_type", "is_business", "date"] if col in client.columns]
if merge_check_cols:
    print(client[merge_check_cols].drop_duplicates().shape)
    display(client[merge_check_cols].drop_duplicates().head(10))
```

```python
## weather mapping 방식 확인
weather_map = pd.read_csv(DATA_DIR / "weather_station_to_county_mapping.csv")

print("weather_map columns:")
print(weather_map.columns.tolist())

display(weather_map.head())

print("\nweather_map 결측치 비율:")
display((weather_map.isnull().mean() * 100).sort_values(ascending=False))

print("\nweather_map 중복 수:")
print(weather_map.duplicated().sum())
```

```python
print("historical_weather columns:")
print(historical_weather.columns.tolist())

print("\nforecast_weather columns:")
print(forecast_weather.columns.tolist())

hist_coord_cols = [col for col in ["latitude", "longitude"] if col in historical_weather.columns]
fcst_coord_cols = [col for col in ["latitude", "longitude"] if col in forecast_weather.columns]

print("\nhistorical_weather 좌표 샘플:")
if hist_coord_cols:
    display(historical_weather[hist_coord_cols].drop_duplicates().head())
else:
    print("historical_weather에 latitude/longitude 없음")

print("\nforecast_weather 좌표 샘플:")
if fcst_coord_cols:
    display(forecast_weather[fcst_coord_cols].drop_duplicates().head())
else:
    print("forecast_weather에 latitude/longitude 없음")
```

```python
## datetime 단위 확인
datetime_candidates = {
    "train": ["datetime"],
    "client": ["date"],
    "electricity_prices": ["forecast_date"],
    "gas_prices": ["forecast_date"],
    "forecast_weather": ["forecast_datetime"],
    "historical_weather": ["datetime"],
}

datasets = {
    "train": train,
    "client": client,
    "electricity_prices": electricity_prices,
    "gas_prices": gas_prices,
    "forecast_weather": forecast_weather,
    "historical_weather": historical_weather,
}

for name, cols in datetime_candidates.items():
    df = datasets[name]
    print(f"\n[{name}]")
    for col in cols:
        if col in df.columns:
            print(f"- {col}")
            print("  dtype:", df[col].dtype)
            print("  min :", df[col].min())
            print("  max :", df[col].max())
            display(df[[col]].head())
        else:
            print(f"- {col} 없음")
```

```python
## group 개수 확인
group_cols = ["county", "product_type", "is_business", "is_consumption"]
existing_group_cols = [col for col in group_cols if col in train.columns]

print("사용 가능한 group cols:")
print(existing_group_cols)

group_count = train[existing_group_cols].drop_duplicates().shape[0]
print("\n고유 group 개수:", group_count)

print("\ngroup 샘플:")
display(train[existing_group_cols].drop_duplicates().head(20))

print("\n각 county별 group 수:")
if "county" in existing_group_cols:
    county_group_count = (
        train[existing_group_cols]
        .drop_duplicates()
        .groupby("county")
        .size()
        .sort_values(ascending=False)
    )
    display(county_group_count.head(20))
```

```python
## 시간 연속성 확인
required_cols = ["county", "product_type", "is_business", "is_consumption", "datetime"]
missing_required = [col for col in required_cols if col not in train.columns]

if missing_required:
    print("다음 컬럼이 없어 시간 연속성 체크를 할 수 없음:", missing_required)
else:
    train_sorted = train.sort_values(required_cols).copy()

    train_sorted["time_diff"] = (
        train_sorted
        .groupby(["county", "product_type", "is_business", "is_consumption"])["datetime"]
        .diff()
    )

    print("group 내부 시간 간격 분포 상위 10개:")
    print(train_sorted["time_diff"].value_counts(dropna=False).head(10))

    print("\n1시간이 아닌 간격 샘플:")
    abnormal_diff = train_sorted[
        train_sorted["time_diff"].notna() &
        (train_sorted["time_diff"] != pd.Timedelta(hours=1))
    ]

    print("1시간이 아닌 행 수:", len(abnormal_diff))
    display(abnormal_diff.head(20))
```

```python
## target 분포 확인
import matplotlib.pyplot as plt

if "target" in train.columns:
    print("전체 target 기술통계:")
    display(train["target"].describe())

    print("\n음수 target 개수:")
    print((train["target"] < 0).sum())

    print("\n0 target 개수:")
    print((train["target"] == 0).sum())

    plt.figure(figsize=(10, 4))
    train["target"].hist(bins=100)
    plt.title("Target Distribution")
    plt.xlabel("target")
    plt.ylabel("count")
    plt.show()

    if "is_consumption" in train.columns:
        print("\nis_consumption별 target 기술통계:")
        display(train.groupby("is_consumption")["target"].describe())

        plt.figure(figsize=(10, 4))
        train.loc[train["is_consumption"] == 1, "target"].hist(bins=100, alpha=0.6, label="consumption")
        train.loc[train["is_consumption"] == 0, "target"].hist(bins=100, alpha=0.6, label="production")
        plt.title("Target Distribution by is_consumption")
        plt.xlabel("target")
        plt.ylabel("count")
        plt.legend()
        plt.show()
else:
    print("train에 target 컬럼이 없음")
```
